```python
# simulate SImodel with various beta 
SI1 = Compartment_Model.SImodel(100, 0.3, initI=10)
SI2 = Compartment_Model.SImodel(100, 0.5, initI=10)
SI3 = Compartment_Model.SImodel(100, 0.8, initI=10)

SI1.run(100)
SI2.run(100)
SI3.run(100)

fig, ax = plt.subplots(figsize = (12,8))

ax.plot(SI1.tseries, SI1.numI / SI1.N, label = "beta = 0.3")
ax.plot(SI2.tseries, SI2.numI / SI2.N, label = "beta = 0.5")
ax.plot(SI3.tseries, SI3.numI / SI3.N, label = "beta = 0.8")

ax.set_xlabel('simulation time', fontsize = 30)
ax.set_ylabel('percent of infected population', fontsize = 25)
ax.set_xlim(0, (max(SI1.tseries)))
ax.set_ylim(0, max(SI3.numI/SI3.N)+0.05)
ax.set_xticklabels(['{:.0f}'.format(x) for x in ax.get_xticks()],fontsize = 25)
ax.set_yticklabels(['{:,.0%}'.format(y) if y != 0 else "" for y in ax.get_yticks()],fontsize = 25)

legend_handles, legend_labels = ax.get_legend_handles_labels()
ax.legend(legend_handles[::-1], legend_labels[::-1], loc='upper right', facecolor='white', edgecolor='none', framealpha=0.9, prop={'size': 15})
ax.set_title("SI model", size=40)

plt.show()
```

```python
# simulate SISmodel with rate of beta and gamma

final_infected = []
for i in np.arange(0,2,0.01):
    SIS = Compartment_Model.SISmodel(100, 0.5*i, 0.5, initI=10)
    SIS.run(1000)
    final_infected.append(SIS.numI[-1]/SIS.N)

fig, ax = plt.subplots(figsize = (12,8))

ax.plot(np.arange(0,2,0.01), final_infected)

ax.set_xlabel('$ \\beta / \gamma $', fontsize = 30)
ax.set_ylabel('percent of infected population', fontsize = 25)
ax.set_xlim(0,2.01)
ax.set_ylim(0, max(final_infected)+0.05)
ax.set_xticklabels(['{:,.2f}'.format(x) for x in ax.get_xticks()],fontsize = 25)
ax.set_yticklabels(['{:,.0%}'.format(y) if y != 0 else "" for y in ax.get_yticks()],fontsize = 25)

legend_handles, legend_labels = ax.get_legend_handles_labels()
ax.legend(legend_handles[::-1], legend_labels[::-1], loc='upper right', facecolor='white', edgecolor='none', framealpha=0.9, prop={'size': 15})
ax.set_title("SIS model", size=40)

plt.show()
```

```python
# simulate SISmodel with rate of beta and gamma

SIS = Compartment_Model.SISmodel(100, 0.5, 0.3, initI=10)
SIS.run(100)

fig, ax = plt.subplots(figsize = (12,8))

ax.plot(SIS.tseries, (SIS.numS/SIS.N)*(SIS.beta/SIS.gamma))

ax.set_xlabel('simulation time', fontsize = 30)
ax.set_ylabel('$ R_e $', fontsize = 25)
ax.set_xlim(0, (max(SIS.tseries)))
ax.set_ylim(min((SIS.numS/SIS.N)*(SIS.beta/SIS.gamma))-0.1, max((SIS.numS/SIS.N)*(SIS.beta/SIS.gamma))+0.1)
ax.set_xticklabels(['{:,.0f}'.format(x) for x in ax.get_xticks()],fontsize = 25)
ax.set_yticklabels(['{:,.2f}'.format(y) if y != 0 else "" for y in ax.get_yticks()],fontsize = 25)

legend_handles, legend_labels = ax.get_legend_handles_labels()
ax.legend(legend_handles[::-1], legend_labels[::-1], loc='upper right', facecolor='white', edgecolor='none', framealpha=0.9, prop={'size': 15})
ax.set_title("SIS model", size=40)

plt.show()
```

# Simulating SEIR by agent-based model

```python
%matplotlib inline

import os
import sys
sys.path.append("/root/.jupyter/python_file/")

from PyCX import pycxsimulator
from pylab import *

import copy as cp

S_init,I_init = 90, 10
ms, mi = 0.03, 0.03
beta = 0.3

# nr = 500. # carrying capacity of rabbits
#
# r_init = 100 # initial rabbit population
# mr = 0.03 # magnitude of movement of rabbits
# dr = 1.0 # death rate of rabbits when it faces foxes
# rr = 0.1 # reproduction rate of rabbits
#
# f_init = 30 # initial fox population
# mf = 0.05 # magnitude of movement of foxes
# df = 0.1 # death rate of foxes when there is no food
# rf = 0.5 # reproduction rate of foxes

cd = 0.02 # radius for collision detection
cdsq = cd ** 2

class agent:
    pass

def initialize():
    global agents
    agents = []
    for i in range(S_init + I_init):
        ag = agent()
        ag.type = 'S' if i < I_init else 'I'
        ag.x = random()
        ag.y = random()
        agents.append(ag)

def observe():
    global agents
    cla()
    S = [ag for ag in agents if ag.type == 'S']
    if len(S) > 0:
        x = [ag.x for ag in S]
        y = [ag.y for ag in S]
        plot(x, y, 'b.')
    I = [ag for ag in agents if ag.type == 'I']
    if len(I) > 0:
        x = [ag.x for ag in I]
        y = [ag.y for ag in I]
        plot(x, y, 'ro')
    axis('image')
    axis([0, 1, 0, 1])

def update_one_agent():
    global agents
    if agents == []:
        return

    ag = choice(agents)

    # simulating random movement
    m = ms if ag.type == 'S' else mi
    ag.x += uniform(-m, m)
    ag.y += uniform(-m, m)
    ag.x = 1 if ag.x > 1 else 0 if ag.x < 0 else ag.x
    ag.y = 1 if ag.y > 1 else 0 if ag.y < 0 else ag.y

    # detecting collision and simulating death or birth
    neighbors = [nb for nb in agents if nb.type != ag.type
                 and (ag.x - nb.x)**2 + (ag.y - nb.y)**2 < cdsq]

    if ag.type == 'S':
        if len(neighbors) > 0: # if there are foxes nearby
            if random() < beta:
                ag.type = 'I'
                return
    # else:
    #     if len(neighbors) == 0: # if there are no rabbits nearby
    #         if random() < df:
    #             agents.remove(ag)
    #             return
    #     else: # if there are rabbits nearby
    #         if random() < rf:
    #             agents.append(cp.copy(ag))

def update():
    global agents
    t = 0.
    while t < 1. and len(agents) > 0:
        t += 1. / len(agents)
        update_one_agent()

pycxsimulator.GUI().start(func=[initialize, observe, update])
```
