# Simulating SI by differential equation

```python
# import model package
import Compartment_Model
import numpy as np

# import plot package
import matplotlib.pyplot as plt
```

```python
# simulate SImodel with line plot
SI = Compartment_Model.SImodel(100, 0.5, initI=1)
SI.run(100)
SI.plot("SI model", S_plot = "line", I_plot = "line")
```

```python
# simulate SImodel with stacked plot
SI = Compartment_Model.SImodel(100, 0.5, initI=1)
SI.run(100)
SI.plot("SI model", S_plot = "stacked", I_plot = "stacked")
```

```python
# simulate SImodel with various beta 
SI1 = Compartment_Model.SImodel(100, 0.3, initI=10)
SI2 = Compartment_Model.SImodel(100, 0.5, initI=10)
SI3 = Compartment_Model.SImodel(100, 0.8, initI=10)

SI1.run(100)
SI2.run(100)
SI3.run(100)

fig, ax = plt.subplots(figsize = (12,8))

ax.plot(SI1.tseries, SI1.numI / SI1.N, label = "beta = 0.3")
ax.plot(SI2.tseries, SI2.numI / SI2.N, label = "beta = 0.5")
ax.plot(SI3.tseries, SI3.numI / SI3.N, label = "beta = 0.8")

ax.set_xlabel('simulation time', fontsize = 30)
ax.set_ylabel('percent of infected population', fontsize = 25)
ax.set_xlim(0, (max(SI1.tseries)))
ax.set_ylim(0, max(SI3.numI/SI3.N)+0.05)
ax.set_xticklabels(['{:.0f}'.format(x) for x in ax.get_xticks()],fontsize = 25)
ax.set_yticklabels(['{:,.0%}'.format(y) if y != 0 else "" for y in ax.get_yticks()],fontsize = 25)

legend_handles, legend_labels = ax.get_legend_handles_labels()
ax.legend(legend_handles[::-1], legend_labels[::-1], loc='upper right', facecolor='white', edgecolor='none', framealpha=0.9, prop={'size': 15})
ax.set_title("SI model", size=40)

plt.show()
```

```python
# simulate SISmodel with various beta 
SIS1 = Compartment_Model.SISmodel(100, 0.5, 0.1, initI=10)
SIS2 = Compartment_Model.SISmodel(100, 0.5, 0.3, initI=10)
SIS3 = Compartment_Model.SISmodel(100, 0.5, 0.5, initI=10)
SIS4 = Compartment_Model.SISmodel(100, 0.5, 0.7, initI=10)
SIS5 = Compartment_Model.SISmodel(100, 0.5, 0.9, initI=10)

SIS1.run(100)
SIS2.run(100)
SIS3.run(100)
SIS4.run(100)
SIS5.run(100)


fig, ax = plt.subplots(figsize = (12,8))

ax.plot(SIS1.tseries, SIS1.numI / SIS1.N, label = "gamma = 0.1")
ax.plot(SIS2.tseries, SIS2.numI / SIS2.N, label = "gamma = 0.3")
ax.plot(SIS3.tseries, SIS3.numI / SIS3.N, label = "gamma = 0.5")
ax.plot(SIS4.tseries, SIS4.numI / SIS4.N, label = "gamma = 0.7")
ax.plot(SIS5.tseries, SIS5.numI / SIS5.N, label = "gamma = 0.9")

ax.set_xlabel('simulation time', fontsize = 30)
ax.set_ylabel('percent of infected population', fontsize = 25)
ax.set_xlim(0, (max(SIS1.tseries)))
ax.set_ylim(0, max(SIS1.numI/SIS1.N)+0.05)
ax.set_xticklabels(['{:.0f}'.format(x) for x in ax.get_xticks()],fontsize = 25)
ax.set_yticklabels(['{:,.0%}'.format(y) if y != 0 else "" for y in ax.get_yticks()],fontsize = 25)

legend_handles, legend_labels = ax.get_legend_handles_labels()
ax.legend(legend_handles[::-1], legend_labels[::-1], loc='upper right', facecolor='white', edgecolor='none', framealpha=0.9, prop={'size': 15})
ax.set_title("SIS model", size=40)

plt.show()
```

```python
# simulate SISmodel with rate of beta and gamma

final_infected = []
for i in np.arange(0,2,0.01):
    SIS = Compartment_Model.SISmodel(100, 0.5*i, 0.5, initI=10)
    SIS.run(1000)
    final_infected.append(SIS.numI[-1]/SIS.N)

fig, ax = plt.subplots(figsize = (12,8))

ax.plot(np.arange(0,2,0.01), final_infected)

ax.set_xlabel('$ \\beta / \gamma $', fontsize = 30)
ax.set_ylabel('percent of infected population', fontsize = 25)
ax.set_xlim(0,2.01)
ax.set_ylim(0, max(final_infected)+0.05)
ax.set_xticklabels(['{:,.2f}'.format(x) for x in ax.get_xticks()],fontsize = 25)
ax.set_yticklabels(['{:,.0%}'.format(y) if y != 0 else "" for y in ax.get_yticks()],fontsize = 25)

legend_handles, legend_labels = ax.get_legend_handles_labels()
ax.legend(legend_handles[::-1], legend_labels[::-1], loc='upper right', facecolor='white', edgecolor='none', framealpha=0.9, prop={'size': 15})
ax.set_title("SIS model", size=40)

plt.show()
```

```python
# simulate SISmodel with rate of beta and gamma

SIS = Compartment_Model.SISmodel(100, 0.5, 0.3, initI=10)
SIS.run(100)

fig, ax = plt.subplots(figsize = (12,8))

ax.plot(SIS.tseries, (SIS.numS/SIS.N)*(SIS.beta/SIS.gamma))

ax.set_xlabel('simulation time', fontsize = 30)
ax.set_ylabel('$ R_e $', fontsize = 25)
ax.set_xlim(0, (max(SIS.tseries)))
ax.set_ylim(min((SIS.numS/SIS.N)*(SIS.beta/SIS.gamma))-0.1, max((SIS.numS/SIS.N)*(SIS.beta/SIS.gamma))+0.1)
ax.set_xticklabels(['{:,.0f}'.format(x) for x in ax.get_xticks()],fontsize = 25)
ax.set_yticklabels(['{:,.2f}'.format(y) if y != 0 else "" for y in ax.get_yticks()],fontsize = 25)

legend_handles, legend_labels = ax.get_legend_handles_labels()
ax.legend(legend_handles[::-1], legend_labels[::-1], loc='upper right', facecolor='white', edgecolor='none', framealpha=0.9, prop={'size': 15})
ax.set_title("SIS model", size=40)

plt.show()
```
